```python
import pandas as pd

PRT_path = 'pf_data_ew.csv'

PriceTrends_df = pd.read_csv(PRT_path, index_col=0)
PriceTrends_df

# 1부터 8까지 -> PriceTrends_{2~9}로 컬럼명 변경
PriceTrends_Low = PriceTrends_df.iloc[:, 0]
PriceTrends_High = PriceTrends_df.iloc[:, 9]
PriceTrends_HL = PriceTrends_df.iloc[:, 10]

PriceTrends_df = PriceTrends_df.iloc[:, 1:9]
PriceTrends_df.columns = ['PriceTrends_' + str(i) for i in range(2, 10)]

# Low, High, HL 컬럼 추가
PriceTrends_df['PriceTrends_Low'] = PriceTrends_Low
PriceTrends_df['PriceTrends_High'] = PriceTrends_High
PriceTrends_df['PriceTrends_HL'] = PriceTrends_HL
PriceTrends_df
```

```python
# PriceTrends_df는 수익률 데이터임.
# 수익률 데이터를 가격으로 변환하고 싶음.
# 가격으로 변환하려면 초기값이 필요함.

# 초기값은 100으로 설정하고, 수익률을 누적하여 곱해주면 됨.
# 누적곱은 cumprod()로 계산 가능.

# 초기값과 누적곱의 수식에 유의하여 다시 계산

# 데이터는 뺴고 칼럼명 데이터프레임 복사
PriceTrendsPrice_df = PriceTrends_df.copy()
PriceTrendsPrice_df = (PriceTrendsPrice_df + 1).cumprod() * 100
PriceTrendsPrice_df.to_csv('PriceTrendsPrice.csv')
```

```python
main_df = pd.read_csv('/home/indi/codespace/PriceTrend/LAA/df_asset_laa.csv', index_col=0)
# DATE를 기준으로 PriceTrendsPrice_df와 main_df를 합침
# 결측치를 그대로 둠
# 기존 DATE가 사라지면 안되므로 left_on, right_index를 사용
# DATE가 사라지면 안되므로 how='left'를 사용
# main_df에 PriceTrendsPrice_df를 합침
main_df = main_df.merge(PriceTrendsPrice_df, left_on='DATE', right_index=True, how='left')

# 결측치는 앞뒤로 채워줌
main_df.fillna(method='ffill', inplace=True)
main_df.fillna(method='bfill', inplace=True)

# 다시 PriceTrendsPrice_df 부분만 빼서 저장
PriceTrendsPrice_df2 = main_df.iloc[:, -11:]

# index는 DATE로 설정
PriceTrendsPrice_df2.index = main_df['DATE']
PriceTrendsPrice_df2.to_csv('PriceTrendsPrice2.csv')
```
